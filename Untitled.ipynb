```python
import pandas as pd 
import geopy
```

```python
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut
from geopy.extra.rate_limiter import RateLimiter
```

```python
df = pd.read_csv("C:\\Users\\chand\\Downloads\\Hot Chocolate 15K Results.csv")
df.head()
```

```python
gender_mapping = {'Male': 'M', 'Female': 'F'}
df['GENDER'] = df['GENDER'].map(gender_mapping)
df.head()
```

```python
df[['CITY', 'STATE']] = df['HOMETOWN'].str.split(', ', expand=True)
df.head()
```

```python
df.dropna()
```

```python
import datetime
def convert_to_time(value):
    time_components = value.split(':')
    hours = int(time_components[0])
    minutes = int(time_components[1])
    seconds = int(time_components[2]) #if len(time_components) == 3 else 0  # If seconds not provided, set to 0
    # Check if seconds are "00"
    if seconds == 0:
        hours, minutes, seconds = seconds, hours, minutes
    return datetime.time(hours, minutes, seconds)
```

```python
for value in df['TIME']:
    convert_to_time(value)
```

```python
df['OFFICIAL TIME'] = df['TIME'].apply(convert_to_time)
```

```python
def calcminutes(time_obj):
  return time_obj.hour * 60 + time_obj.minute + time_obj.second / 60
df['TotalMinutes'] = df['OFFICIAL TIME'].apply(calcminutes)
df['TotalMinutes'] = df['TotalMinutes'].round().astype(int)
```

```python
df.head()
```

```python
def get_lat_long(CITY, STATE):
    address = f"{CITY}, {STATE}"
    try:
        geolocator = Nominatim(user_agent="running", timeout=15)
        location = geolocator.geocode(address)
        if location:
            return location.latitude, location.longitude
        else:
            return None, None
    except GeocoderTimedOut:
        return None, None
    
```

```python
df['LATITUDE'], df['LONGITUDE'] = zip(*df.apply(lambda x: get_lat_long(x['CITY'], x['STATE']), axis=1))
```

```python
df['LATLONG'] = df['LATITUDE'].astype(str)+','+ df['LONGITUDE'].astype(str)
```
